#**Data Preprocessing**

**Import Libraries**

```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder
```

**Load Dataset**

```python
df = pd.read_csv("../data/credit_risk_dataset.csv")
data = df.copy()
```

**Handling Missing Values**

Missing values are handled by replacing them with the median:

```python
data['person_emp_length'].fillna(data['person_emp_length'].median(), inplace=True)
data['loan_int_rate'].fillna(data['loan_int_rate'].median(), inplace=True)
```

**Outlier Handling**

To reduce the impact of extreme values, we apply capping using quantiles (1% and 99%).

```python
def cap_outliers(df, col):
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = np.clip(df[col], lower, upper)

outlier_cols = ['person_age', 'person_emp_length']

for col in outlier_cols:
    cap_outliers(data, col)
```

**Encoding Categorical Variables**

```python
le = LabelEncoder()
data['loan_grade'] = le.fit_transform(data['loan_grade'])
data['cb_person_default_on_file'] = le.fit_transform(data['cb_person_default_on_file'])
```

```python
data = pd.get_dummies(
    data,
    columns=['person_home_ownership', 'loan_intent'],
    drop_first=True
)
bool_cols = data.select_dtypes(include=['bool']).columns
data[bool_cols] = data[bool_cols].astype(int)
```

**Save clean data**

```python
data.to_csv("../data/credit_risk_clean.csv", index=False)
```
